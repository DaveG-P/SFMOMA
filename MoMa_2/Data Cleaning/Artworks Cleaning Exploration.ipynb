```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/new_artworks.csv"
```

```python
artworks_df = pd.read_csv(path)
artworks_df
```

```python
# Checking Nan values
artworks_df.isnull().values.any()
```

```python
# Dropping Object Number
# df.drop(columns=['B', 'C'])
AW = artworks_df.drop(columns=["Object Number", "Dimensions"]).set_index("Name")
AW.tail(50)
```

```python
# Updated CSV 
AW.to_csv("Artworks.csv", sep=",")
```
