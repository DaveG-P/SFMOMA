```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Data/Old_Data/new_artists.csv"
```

```python
artists_df = pd.read_csv(path)
```

```python
# Old Data Frame
artists_df
```

```python
artists_df.drop(artists_df.columns[[0]], axis=1, inplace=True)
artists_df
```

```python
# Setting Index to Name
artists_df.set_index('Name')
```

```python
artists_df.count()
```

```python
artists_df = artists_df.dropna(how='any')
artists_df.count()
```

```python
# Resetting index in new data frame
artists_df
```

```python
artists_df.dtypes
```

```python
# Changing Death Year Column in to "Living Status"
 # Using .rename(columns={}) in order to rename columns
artists_df['Living Status'] = artists_df['Death Year'] == 'Alive'
# Setting Index to Name
artists_df
```

```python
# Setting Index to Name
artists_df.set_index('Name')
```

```python
# ---- REMEBER ME ----
# Updated csv file
artists_df.to_csv("Artists.csv", sep=",")
```

## Analysing Nationalies

```python
Nationality_df = pd.DataFrame(artists_df["Nationality"].value_counts())
Nationality_df
```

```python
# Nationaloty Count and deleted values counted as 1
Nationality_df = pd.DataFrame(artists_df["Nationality"].value_counts())[: -60]

Nationality_df
```

```python
# Created new Nationality csv file
Nationality_df.to_csv("Nationality.csv", sep=",")
```

```python
# Getting total count of Nationalities
# Getting Percentage of countries  represented in MoMa
nations = Nationality_df["Nationality"].count()/58 * 100
nations
```

##  Data Visualization

```python
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import matplotlib.cm as cm
import numpy as np


# Data are 256x256 16 bit integers
with cbook.get_sample_data('s1045.ima.gz') as dfile:
    im = np.fromstring(dfile.read(), np.uint16).reshape((256, 256))

fig, ax = plt.subplots(num="MRI_demo")
ax.imshow(im, cmap=cm.gray)
ax.axis('off')

plt.show()
```
