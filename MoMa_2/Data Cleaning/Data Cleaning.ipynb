```python
import pandas as pd
import numpy as np
```

```python
file_one = "Data/museum-collection/artists.csv"
file_two = "Data/museum-collection/artworks.csv"
```

```python
artists = pd.read_csv(file_one)
```

```python
artists.head(50)
```

```python
# Replacing Nan in Death Year with Alive status
artists["Death Year"] = artists["Death Year"].fillna("Alive")
artists
```

```python
# Setting the index to the Name of the Artist
artists.set_index("Name")
```

```python
artists.to_csv("new_artists.csv", sep=",")
```

```python
# Cleaning Artworks csv
file_two_df = pd.read_csv(file_two)
file_two_df.head()
```

```python
# Removing unneccesary columns
artworks = file_two_df[["Artwork ID", "Title", "Artist ID", "Name", "Date", 
                       "Medium", "Dimensions", "Acquisition Date", "Classification", 
                       "Object Number"]]
artworks.head()
```

```python
# Data Cleaning
artworks.count()
```

```python
artworks = artworks.dropna(how='any')
artworks.count()
```

```python
artworks = artworks.set_index('Name')
artworks.head()
```

```python
artworks.to_csv("new_artworks.csv", sep=",")
```
